```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load the seaborn data set diamonds
df = sns.load_dataset('diamonds')
```

```python
df = df.sample(frac=0.10,random_state=33)
```

```python
df.head()
```

```python
sns.relplot(x='price',y='carat',data=df)
```

```python
#Box plot in matplotlib
plt.boxplot(df['price'])
```

```python
sns.jointplot(x='price',y='carat',data=df)
```

```python
sns.pairplot(df)
```

```python
sns.catplot(data=df,x='color',y='price',kind='box')
```

```python
sns.catplot(data=df,x='color',y='price',col='cut',kind='box')
```

```python
df['premium'] = df['cut'].apply(lambda x:True if x=='Premium' else False)
```

```python
sns.catplot(data=df,x='color',y='price',hue ='premium',kind='box')
```

```python
sns.barplot(data=df,x='color',y='price',errorbar=None)
```

```python
sns.barplot(data=df.sort_values('price'),x='color',y='price',errorbar=None,palette='viridis')
plt.show()
```

```python
sns.color_palette()
```

```python
#Violin Plots
sns.catplot(data=df,x='cut',y='price',kind='violin')
```

```python
sns.catplot(data=df,x='cut',y='price',kind='swarm')
```

```python
sns.catplot(data=df,x='cut',y='price',kind='strip')
```

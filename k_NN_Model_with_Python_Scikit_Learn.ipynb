```python
#import pandas and sklearn
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Read the data set
df = pd.read_csv("diabetes_data.csv")
```

```python
#create a new data frame with out the target variable
X = df.drop(columns = ["diabetes"])
```

```python
type(X)
```

```python
X.head()
```

```python
#create a vector array variable to represent the target variable
y = df['diabetes'].values
```

```python
#split the data into train and test data sets using sklearn
#inputs - nXp matrix of dependent variables, a vector with target data set, test _size = 20% data for test
# and 80% for training, random_sate = 1 - we get the same split each time of randomeny selecting
# dealing with imbalanced datasets to ensure that the training and test datasets
# have the same proportion of class labels as the input dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 1, stratify = y)
```

```python
X_train
```

```python
df.iloc[665]
```

```python
y_train
```

```python
#build the model using K-nn classifer algorithm
# create K-nn classifier

knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
type(knn)
```

```python
type(X_train)
```

```python
# Fit the classifier to the data
knn.fit(X_train,y_train)
```

```python
#show first 5 model predictions on the test data
knn.predict(X_test)
```

```python
#check accuracy of our model on the test data
knn.score(X_test, y_test)
```

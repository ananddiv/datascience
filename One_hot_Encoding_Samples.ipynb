```python
import pandas as pd
```

```python
df = pd.DataFrame({"name":["William","Emma","Sofia","Markus","Edward","Thomas","Ethan","Olivia","Arun","Anika","Paulo"],
 "sales":[50000,52000,90000,34000,42000,72000,49000,55000,67000,65000,67000],
 "region":["East","North","East","South","West","West","South","West","West","East",'North']})
```

```python
df
```

```python
df = pd.get_dummies(data=df,columns=['region'],dtype=int)
```

```python
df = pd.get_dummies(data=df,columns=['region'],dtype=int,drop_first = True)
```

```python
df
```

```python
df_regions = pd.get_dummies(df['region'],dtype=int,drop_first=True)
```

```python
df_regions
```

```python
df_wdummy = pd.concat([df,df_regions],axis=1)
```

```python
df_wdummy
```

# Observation[year_birth]
Outliers found in year_birth feature.
Options:
1. Ignore these rows/data points
2. Set the birth year to the median birth year.
3. Set the birth year to the minumum valid value - 1940


```python
#View outlier date for year_birth <1920
df_clean[df_clean['year_birth'] < 1920]
```

```python
#Find the median year_birth
median_year_birth = df_clean[df_clean['year_birth'] >=1920]['year_birth'].median()
```

```python
#Replace the year_birth outliers with the median year
df_clean[df_clean['year_birth']<1920]= df_clean[df_clean['year_birth']<1920].assign(year_birth = median_year_birth)
```

```python
#View outlier date for year_birth <1920
df_clean[df_clean['year_birth'] < 1920]
```

```python
#Analysing the year_birth column after removing the outliers.
sns.histplot(df_clean['year_birth'])
```
